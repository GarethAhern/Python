```python
#importing matplotlib to plot the graphs
import matplotlib.pyplot as plt

#to avoid pop-ups & show graphs inline with the code
%matplotlib inline

#pandas is required to read the input dataset
import pandas as pd

import numpy as np
```

weâ€™ll be using Figure, Axes together using plt.subplots() function just because this is the most used way.

Figure constitutes of subplots, sub axis, titles, subtitles, legends, everything inside the plot but an overview.

Axes define a subplot, we can write our own x-axis limits, y-axis limits, their labels, the type of graph. It controls every detail inside the subplot.

```python
#data
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
x2 = x*2 # You can multiply a whole list by 2 if it's an np.array
y2 = y*2



# Create a variable 'fig' and 'ax'
fig, ax = plt.subplots(1,2) # 1 row, 2 columns 

#fig helps control the overall size of the combination of plots. 

ax[0].plot(x,y) # ax 0 refers to the first plot 
ax[1].plot(y,x) # ax 1 refers to the second
plt.show() # show the plot 
```

```python
fig, ax = plt.subplots(2,2) # If you wanted 2x2 plots (2 in row, 2 in column )

#bottom left 
ax[1,0].plot(x,x2)

#bottom right 
ax[1,1].plot(x,y2)
plt.show()
```
