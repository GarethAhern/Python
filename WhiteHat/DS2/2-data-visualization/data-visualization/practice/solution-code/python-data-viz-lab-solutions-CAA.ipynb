```python
# Get rid of 'Null' values 
df = df.replace('Null',np.NaN)
df
```

### Convert pax_percent fields into floats.

```python
#df = df.astype({'total_pax_percent': 'float64'}).dtypes

columns_to_convert = {'total_pax_percent':'float64',
                     'terminal_pax_percent':'float64',
                     'transit_pax_percent':'float64'}
df=df.astype(columns_to_convert)
df.dtypes
```

```python
df['total_pax_percent'] = df['total_pax_percent'].round(2)
df['terminal_pax_percent'] = df['terminal_pax_percent'].round(2)
df['transit_pax_percent'] = df['transit_pax_percent'].round(2)
df
```

```python
reporting = df['reporting_airport_group_name']

other_airports = reporting == 'Other UK Airports'
other_airports_df = df[other_airports]
other_airports_df.head(10)
```

## Data Visualisation

```python
#Create Dataframes for the other groups 
london_airports = reporting == 'London Area Airports'
london_airports_df = df[london_airports]

non_uk_airports =reporting == 'Non UK Reporting Airports'
non_uk_airports_df = df[non_uk_airports]
```

```python
# Can also be done in sub plots or seperate cells or with a for loop 


#below is subplot method with forloop 
#Create list with dataframes 
df_list =[other_airports_df, london_airports_df, non_uk_airports_df]

#Set rows and columns for subplots 
ncol = 1
nrow=3

index = 0


fig, axes = plt.subplots(nrow, ncol) #Initilise subplot

for r in range(nrow):
    for c in range(ncol):
        df_list[index].plot(
            x='reporting_airport_name', # Select reporting airport column as x axis
            y =['total_pax_last_period','total_pax_this_period'], # Use data from this/last period
            kind='barh', # Make into horizontal bar graph
            xlim=(0,55000), # Change the x scale from 0 -55000
            ax=axes[r,c]
        )
        index +=1

```
